## Creating a Copy of the dataset

### AgeGroup Column:

Insights:
1. Upon checking the percentage of missing values for each column in the dataset, we decided to drop the rows with missing values for the 'AgeGroup' Column because it made up less than 2% of the dataset.

```python
#drop missing values in the AgeGroup column.
batch4_manila_copy.dropna(subset=['AgeGroup'], inplace=True)
```

```python
batch4_manila_copy.head()
```

```python
missing_batch4_manila_copy = batch4_manila_copy.isnull().sum().to_frame().rename(columns={0:"Total No. of Missing Values"})


# Divided the total No. of Missing values by the Total length of the dataframe, then round it to 2 decimal places
missing_batch4_manila_copy["% of Missing Values"] = round((missing_batch4_manila_copy["Total No. of Missing Values"]/len(batch4_manila_copy))*100,2)





missing_batch4_manila_copy
```

### Check Unique Values of Categorical Attributes.

```python
unq_columns = batch4_manila_copy.select_dtypes(include="O").columns      #selects the columns of data types that are objects



unq_columns
```

```python
#not sure if this is needed, I was just checking notes from the lectures

def preprocess(ValidationStatus):
    ValidationStatus = ValidationStatus.str.replace("(<br/>)", "")
    ValidationStatus = ValidationStatus.str.replace('(<a).*(>).*(</a>)', '')
    ValidationStatus = ValidationStatus.str.replace('(&amp)', '')
    ValidationStatus = ValidationStatus.str.replace('(&gt)', '')
    ValidationStatus = ValidationStatus.str.replace('(&lt)', '')
    ValidationStatus = ValidationStatus.str.replace('(\xa0)', ' ') 
    ValidationStatus = ValidationStatus.str.replace('(\n)', ' ') 
    return ValidationStatus

batch4_manila_copy['ValidationStatus'] = preprocess(batch4_manila_copy['ValidationStatus'])
```
